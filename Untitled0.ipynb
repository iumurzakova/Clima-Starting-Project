```python
import matplotlib.pyplot as plt
import numpy as np

# Circle parameters
R = 5.0  # Radius of the circle
f = 2.0  # Frequency of rotation (in Hz)
v = 1.0  # Translational velocity (in units per second)

# Calculate angular velocity (omega) in radians per second
omega = 2 * np.pi * f

# Time parameters
t_max = 10.0  # Maximum time for simulation (seconds)
dt = 0.01     # Time step

# Initialize arrays to store x and y coordinates of the point's trajectory
x_coords = []
y_coords = []

# Simulate the motion and record the trajectory
t = 0
while t <= t_max:
    # Calculate the angle (theta) based on angular velocity and time
    theta = omega * t

    # Calculate x and y coordinates of the point on the circle
    x = R * np.cos(theta) + v * t
    y = R * np.sin(theta)

    # Append coordinates to the trajectory arrays
    x_coords.append(x)
    y_coords.append(y)

    # Increment time
    t += dt

# Plot the trajectory
plt.figure()
plt.plot(x_coords, y_coords)
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Trajectory of a Point on a Rotating and Translating Circle')
plt.axis('equal')
plt.grid(True)
plt.show()
```
